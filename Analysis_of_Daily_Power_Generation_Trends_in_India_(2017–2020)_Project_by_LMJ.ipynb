<p style="font-size:16px;color:Black;line-height:1.5;text-align:justify"> i) During the data cleaning process, it was observed that 4,667 records had missing values in the `state_code`, `sector`, and `station_type` columns. Further investigation revealed that these records correspond to Bhutan, from which India imports a significant share of hydroelectric power. Since Bhutan is not an Indian state, these missing values do not represent data quality issues. Instead of removing these records, a new column named `country` was created to distinguish domestic power generation (`India`) from imported power (`Bhutan`). This approach preserves all valid observations while incorporating domain knowledge into the analysis.

```python
# Creating new column named "Country"

df['country'] = 'India'
df.loc[df['state_name']=='Bhutan','country'] = 'Bhutan'
```

<p style="font-size:16px;color:Black;line-height:1.5;text-align:justify"> ii) Three records were found to have missing values in the `monitored_capacity`, `todays_gen_prgm`, and `todays_gen_act` columns. Upon further examination, these records were identified as unique observations belonging to a specific power station unit in Maharashtra, and no reliable information was available to impute the missing values. Since the missing records constituted less than 0.001% of the total dataset, they were removed from the analysis to maintain data quality and ensure the accuracy of the results. </p>

<h5 style="text-align:left;color:black;background-color:#E2DCE6;padding:8px;border-radius:8px;font-size:20px;""><b><i>4. Creating Derived columns:</i></b></h5>

<p style="font-size:16px;color:Black;line-height:1.5;text-align:justify">To enhance the analysis, two new columns were created in the dataset.</p>
<ol style="font-size:16px;color:Black;line-height:1.5;text-align:justify"> <li> The first column, power_source, was introduced to distinguish between domestic power generation and imported power. Records belonging to India were classified as Domestic Power, while records related to Bhutan were classified as Imported Power. </li>
<li> The second column, generation_status, was created to compare planned power generation (todays_gen_prgm) with actual power generation (todays_gen_act). Based on predefined conditions, each record was categorized as Exceeded Planned Generation, Below Planned Generation, Met Planned Generation, No Planned Generation, No Actual Generation, or No Planned and No Actual Generation. These derived columns provide better insights into the source of electricity and the performance of power generation against planned targets.</li>
<li> The third derived column is Year, which is extracted from the Date column. This column wiil be useful for analyzing year-wise power generation trends and comparing electricity generation across different years.</li></ol>

```python
# Display the count of each country

df['country'].value_counts()
```

```python
# Creating new coulmn 'Power Source'
df['power_source'] = np.where(
    df['country'] == 'Bhutan',
    'Imported Power',
    'Domestic Power' )
```

```python
# Creating new column generation_status
conditions = [
    # Both are zero
    (df['todays_gen_act'] == 0) & (df['todays_gen_prgm'] == 0),
    
    # Planned generation is zero
    (df['todays_gen_prgm'] == 0),

    # Actual generation is zero
    (df['todays_gen_act'] == 0),

    # Actual > Planned
    (df['todays_gen_act'] > df['todays_gen_prgm']),

    # Actual < Planned
    (df['todays_gen_act'] < df['todays_gen_prgm']),

    # Actual = Planned
    (df['todays_gen_act'] == df['todays_gen_prgm'])

]

choices = [

    'No Planned and No Actual Generation',

    'No Planned Generation',

    'No Actual Generation',

    'Exceeded Planned Generation',

    'Below Planned Generation',

    'Met Planned Generation'

]

df['generation_status'] = np.select(
    conditions,
    choices,
    default='Unknown'
)
```

```python
# Creating new coulmn 'Year'
df['Year'] = df['date'].dt.year
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
```

<h5 style="text-align:left;color:black;background-color:#E2DCE6;padding:8px;border-radius:8px;font-size:20px;""><b><i>1. Month-wise total power generation for every year.</i></b></h5>

<p style="font-size:16px;color:Black;line-height:1.5;text-align:justify"><b>Purpose:</b> To analyze monthly trends and seasonal variations in electricity generation and compare how power generation patterns differ across years. </p>

```python
# Creating month column

df['Month'] = df['Date'].dt.month_name().str[:3]

# Define month order

month_order = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
               'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

df['Month'] = pd.Categorical(
    df['Month'],
    categories=month_order,
    ordered=True
)

# Aggregate month-wise generation

monthly_gen = (
    df.groupby(['Year', 'Month'])['Todays Gen Act (MWh)']
      .sum()
      .reset_index()
)
```

```python
plt.figure(figsize=(14, 7))

sns.lineplot(
    data=monthly_gen,
    x='Month',
    y='Todays Gen Act (MWh)',
    hue='Year',
    marker='o',
    linewidth=2.5
)

plt.title('Month-wise Total Power Generation for Each Year',
          fontsize=16,
          fontweight='bold')

plt.xlabel('Month', fontsize=12)
plt.ylabel('Total Power Generation (MWh)', fontsize=12)
plt.legend(title='Year')
plt.xticks(rotation=45)
plt.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

<p style="font-size:16px;color:Black;line-height:1.5;text-align:justify"><b>Insight:</b> 
<li style="font-size:16px;color:Black;line-height:1.5;text-align:justify"> Power generation exhibits clear seasonal fluctuations, with generation levels varying across different months each year.</li>
<li style="font-size:16px;color:Black;line-height:1.5;text-align:justify"> 2018 and 2019 show relatively stable and consistently high generation, with several months recording generation above 100,000 MWh.</li>
<li style="font-size:16px;color:Black;line-height:1.5;text-align:justify"> May 2019 records the highest monthly generation in the study period, indicating peak electricity demand and production.</li>
<li style="font-size:16px;color:Black;line-height:1.5;text-align:justify">A sharp decline is observed in March 2020, followed by a gradual recovery during the subsequent months.</li>
